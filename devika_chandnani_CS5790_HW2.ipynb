```python
#HomeworkQ2 (40 points) - L2 Linear Regression with Lambda ranging from 0 - 150 
#for each of the 6 datasets, with MSE graphs and least MSE printed
def Graphing(TrainMSQ, TestMSQ,lamdaList, title, trainLegend, testLegend, subplotIndex):
	plt.subplot(subplotIndex)
	plt.plot(lamdaList,TrainMSQ, color="cyan")
	plt.plot(lamdaList,TestMSQ, color="magenta")
	plt.title(title)
	plt.legend([trainLegend, testLegend])
	plt.xlabel('Lambda')
	plt.ylabel('Mean Square Error')
	

def getACCESS(xdata):
	dataACCESS = numpy.insert(xdata,0,1,axis=1)
	return dataACCESS

def WM(mx1, my1, lambdaValue): #figure out W 
	identityMatrix = numpy.identity( numpy.shape(mx1)[1] )
	WMmatrix = (numpy.linalg.inv((mx1.T * mx1) + ( lambdaValue * identityMatrix))) * ( (mx1.T * my1) )
	return WMmatrix

def calcMSE(mx1, my1, WMmatrix): #Calculate MSE
	estimatedYMatrix = mx1 * WMmatrix
	differenceYMatrix = my1 - estimatedYMatrix
	diffSquaredMatrix = numpy.square(differenceYMatrix)
	MSE = numpy.sum(diffSquaredMatrix) / numpy.shape( mx1 )[0]
	return MSE

figure = plt.figure( figsize=(20,10) )

csv_100_10 = ['data/train-100-10.csv','data/test-100-10.csv', 10, 'Dataset 100-10', 'train-100-10', 'test-100-10', 231]
csv_100_100 = ['data/train-100-100.csv' , 'data/test-100-100.csv', 100, 'Dataset 100-100', 'train-100-100', 'test-100-100', 232]
csv_1000_100 = ['data/train-1000-100.csv', 'data/test-1000-100.csv', 100, 'Dataset 1000-100', 'train-1000-100', 'test-1000-100', 233]
csv_50_1000_100 = ['data/train-50(1000)-100.csv', 'data/test-1000-100.csv', 100, 'Dataset 50(1000)-100', 'train-50(1000)-100', 'test-1000-100', 234]
csv_100_1000_100 = ['data/train-100(1000)-100.csv','data/test-1000-100.csv', 100, 'Dataset 100(1000)-10', 'train-100(1000)-100', 'test-1000-100', 235]
csv_150_1000_100 = ['data/train-150(1000)-100.csv','data/test-1000-100.csv', 100, 'Dataset 150(1000)-100', 'train-150(1000)-100', 'test-1000-100', 236]

ALLCSV = [csv_100_10, csv_100_100, csv_1000_100, csv_50_1000_100, csv_100_1000_100, csv_150_1000_100]

print("Answer to Question 2A - Show which Lambda Value gives the least test set MSE")

for item in range(len(ALLCSV)):
	print('\nFinding MSE for', ALLCSV[item][0])

	trainData= numpy.genfromtxt( ALLCSV[item][0], delimiter=',', skip_header=1 )
	testData= numpy.genfromtxt( ALLCSV[item][1], delimiter=',', skip_header=1 )

	trainX = numpy.asmatrix( trainData[:, range(0,ALLCSV[item][2])] )
	trainXACCESS = getACCESS( trainX )
	trainY = numpy.asmatrix( trainData[:, [ALLCSV[item][2]]] )
	testX = numpy.asmatrix( testData[:, range(0,ALLCSV[item][2])] )
	testXACCESS = getACCESS( testX )
	testY = numpy.asmatrix( testData[:, [ALLCSV[item][2]]] )
	
	lambdaList = list(range(0,151))
	mseTrainArray = []
	mseTestArray = []
	wMatrix = None
	
	for in_Item in range(len(lambdaList)):
		WMmatrix = WM( trainXACCESS, trainY, lambdaList[in_Item] )
		mseTrainArray.append( calcMSE(trainXACCESS, trainY, WMmatrix ) )
		mseTestArray.append( calcMSE(testXACCESS, testY, WMmatrix ) )

	print("The least MSE for Test Data Set corresponding to", ALLCSV[item][3], "is", numpy.min(mseTestArray), " - this happens when Lambda is", numpy.argmin(mseTestArray))
	print("The least MSE for Training Data Set corresponding to", ALLCSV[item][3], "is", numpy.min(mseTrainArray), " - this happens when Lambda is", numpy.argmin(mseTrainArray))
	Graphing(mseTrainArray, mseTestArray,lambdaList, ALLCSV[item][3], ALLCSV[item][4], ALLCSV[item][5], ALLCSV[item][6])

plt.subplots_adjust(hspace = 0.5, wspace = 0.5) 
plt.show()
figure.savefig("Ans_Question2.pdf")
	
```

```python
#HomeworkQ3a - Implement 10-Fold Cross Validation and print best choice Lambda and it's corresponding test set MSE.

csv_100_10 = ['data/train-100-10.csv','data/test-100-10.csv', 10, 'Dataset 100-10', 'train-100-10', 'test-100-10', 231]
csv_100_100 = ['data/train-100-100.csv' , 'data/test-100-100.csv', 100, 'Dataset 100-100', 'train-100-100', 'test-100-100', 232]
csv_1000_100 = ['data/train-1000-100.csv', 'data/test-1000-100.csv', 100, 'Dataset 1000-100', 'train-1000-100', 'test-1000-100', 233]
csv_50_1000_100 = ['data/train-50(1000)-100.csv', 'data/test-1000-100.csv', 100, 'Dataset 50(1000)-100', 'train-50(1000)-100', 'test-1000-100', 234]
csv_100_1000_100 = ['data/train-100(1000)-100.csv','data/test-1000-100.csv', 100, 'Dataset 100(1000)-10', 'train-100(1000)-100', 'test-1000-100', 235]
csv_150_1000_100 = ['data/train-150(1000)-100.csv','data/test-1000-100.csv', 100, 'Dataset 150(1000)-100', 'train-150(1000)-100', 'test-1000-100', 236]

ALLCSV = [csv_100_10, csv_100_100, csv_1000_100, csv_50_1000_100, csv_100_1000_100, csv_150_1000_100 ]

tenFCV = 10

for item in range(len(ALLCSV)):
    
    print('\n10-Fold Cross Validation applied to data', ALLCSV[item][0])
    
    trainData= numpy.genfromtxt( ALLCSV[item][0], delimiter=',', skip_header=1 )
    testData= numpy.genfromtxt( ALLCSV[item][1], delimiter=',', skip_header=1 )

    trainX = numpy.asmatrix( trainData[:, range(0,ALLCSV[item][2])] )
    trainXACCESS = getACCESS( trainX )
    trainY = numpy.asmatrix( trainData[:, [ALLCSV[item][2]]] )
    
    testX = numpy.asmatrix( testData[:, range(0,ALLCSV[item][2])] )
    testXACCESS = getACCESS( testX )
    testY = numpy.asmatrix( testData[:, [ALLCSV[item][2]]] )
    
    split_train_data = numpy.split(trainData, tenFCV)

    lambda_one_onefifty = list(range(0,151))
    
    mseTestArray = []
      
    for lambda_val in range(len(lambda_one_onefifty)):
        tenFCV_train_data = None 
        tenFCV_test_data = None
        temp_train_data = None
        
        sum_of_MSE = 0
        # Inner loop to run for 10 fold Cross Validation
        for i in range(len(split_train_data)):

            temp_train_data = numpy.delete(split_train_data, [i], 0)
            tenFCV_test_data = split_train_data[i]
            temp_train_data_shape = numpy.shape( temp_train_data )
            tenFCV_train_data = numpy.reshape(temp_train_data, ( temp_train_data_shape[0] * temp_train_data_shape[1], temp_train_data_shape[2]))
            
            ten_FCV_trainx = numpy.asmatrix( tenFCV_train_data[ : , range(0,ALLCSV[item][2])] )
            ten_FCV_trainXaccess = getACCESS( ten_FCV_trainx )

            ten_FCV_testx = numpy.asmatrix( tenFCV_test_data[ : , range(0,ALLCSV[item][2])] )
            ten_FCV_testXaccess = getACCESS( ten_FCV_testx )
            
            ten_FCV_trainy = numpy.asmatrix( tenFCV_train_data[:, [ALLCSV[item][2]]] )
            ten_FCV_testy = numpy.asmatrix( tenFCV_test_data[:, [ALLCSV[item][2]]] )
            
            ten_FCV_matrix = WM( ten_FCV_trainXaccess, ten_FCV_trainy, lambda_one_onefifty[lambda_val] )
            
            mse = calcMSE(ten_FCV_testXaccess, ten_FCV_testy, ten_FCV_matrix )
            sum_of_MSE += mse
        
        mseTestArray.append([lambda_one_onefifty[lambda_val] , sum_of_MSE/tenFCV])  


    least_MSQ = min(mseTestArray, key=operator.itemgetter(1))

    print( "The least training data MSE using Cross Validation is", least_MSQ[1], "- this happens at Lambda", least_MSQ[0] )
    
    wMatrix = WM( trainXACCESS, trainY, least_MSQ[0] )
    mse = calcMSE(testXACCESS, testY, wMatrix )
    
    print("At this optimal Lambda value of", least_MSQ[0], "- the corresponding test data MSE is", mse)
    
```

```python
#HomeworkQ4 - Learning Curves at Lambda Values = 1, 25, 150 for dataset 1000-100.csv.

def plotMSE(TestMSQ, title, subplotIndex):
    plt.subplot(subplotIndex)
    plt.plot(TestMSQ, color='Magenta')
    plt.title(title)
    plt.xlabel('Size of Training Set')
    plt.ylabel('Mean Square Error')

subplotIndex = 131
figure = plt.figure( figsize=(20,5) )

ALLCSV = [csv_1000_100]

# Number of times to run
drawRandom_subsets = 100
to_repeat = 10

for item in range(len(ALLCSV)):
    
    full_train_data= numpy.genfromtxt( ALLCSV[item][0], delimiter=',', skip_header=1 )
    test_data= numpy.genfromtxt( ALLCSV[item][1], delimiter=',', skip_header=1 )
    
    # Building the training data in the loop. (Random Number loop)

    testX = numpy.asmatrix( test_data[:, range(0,ALLCSV[item][2])] )
    testXACCESS = getACCESS( testX )
    testY = numpy.asmatrix( test_data[:, [ALLCSV[item][2]]] )
    
    lambdaList = list([1,25,150])
    
    for lambdaValue in range(len(lambdaList)):
        
        print("The Learning Curve for when Lambda = ", lambdaList[lambdaValue])
        print("Recording Performances by repeating the process 10 times:")
		
        sum_MSQ = []
        for i in range(0,to_repeat):
            
            list_random = random.sample( range(1,1000), drawRandom_subsets )
            list_random.sort()
            
            list_MSE = []
            for number_random in range(len(list_random) ):

                trainData = full_train_data[:list_random[number_random]]
                trainX = numpy.asmatrix( trainData[:, range(0,ALLCSV[item][2])] )
                trainXaccess = getACCESS( trainX )
                trainY = numpy.asmatrix( trainData[:, [ALLCSV[item][2]]] )
                wMatrix = WM( trainXaccess, trainY, lambdaList[lambdaValue] )
                mse = calcMSE(testXACCESS, testY, wMatrix )
                list_MSE.append( mse )
                
            sum_MSQ += list_MSE
          
        reshape_array_MSE = numpy.reshape( numpy.array(sum_MSQ) , (to_repeat, drawRandom_subsets) )
        avg_array_MSE = numpy.divide( reshape_array_MSE.sum(axis=0) , to_repeat ) 

        print("Average Mean Square Errors for 10 iterations printed below == \n", avg_array_MSE)
        
        plotMSE(avg_array_MSE, "Learning Curve for Lambda: "+str(lambdaList[lambdaValue]), subplotIndex)
        subplotIndex += 1
            
plt.subplots_adjust(wspace = 0.6, hspace = 0.6)
plt.show()

figure.savefig("Ans_Question4.pdf")
```
